### Welcome to Python Notebook! 🐍

**Notebook kya hai?**  
Yeh ek interactive file hai jisme aap code likh sakte ho, turant run kar sakte ho, aur notes bhi likh sakte ho — sab ek hi jagah!

**Kaise chalayein?**  
- Har **cell** (box) ko alag run karo — `Shift + Enter` dabao  
- Markdown cells = notes (yeh wala cell)  
- Code cells = actual Python code

```python
# print() function screen par kuch dikhata hai — sabse pehla Python command!
# Jo bhi quotes ke andar likhoge, woh output mein dikhega
print("hitesh")
```

```python
# for loop = ek kaam baar-baar karna (yahan 5 baar)
# range(5) = 0, 1, 2, 3, 4 (5 tak, 5 shamil nahi)
# i har baar next number hota hai

for i in range(5):
    # f-string (f"...") ke andar {i} ki jagah current value aa jati hai
    print(f"this value is {i}")
```

### 📝 Quick Recap (Yaad rakho)

| Concept | Matlab |
|---------|--------|
| `print()` | Screen par text dikhana |
| `for` loop | Same code multiple times chalana |
| `range(5)` | 0 se 4 tak numbers |
| f-string `f"{i}"` | Variable ki value string mein daalna |
